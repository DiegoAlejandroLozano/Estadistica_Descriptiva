```python
import pandas as pd
from matplotlib import pyplot as plt
```

**Ejemplo 1 - Variable cuantitativa y datos no agrupados.** El númerode hijos en 25 familias es:
 
1, 2, 4, 2, 2, 2, 3, 2, 1, 1, 0, 2, 2, 0, 2, 2, 1, 2, 2, 3, 1, 2, 2, 1, 2

Encontrar la tabla de frecuencias

```python
datos = [1, 2, 4, 2, 2, 2, 3, 2, 1, 1, 0, 2, 2, 0, 2, 2, 1, 2, 2, 3, 1, 2, 2, 1, 2]
num_hijos = pd.Series(datos)
num_hijos.head()
```

```python
tabla_frecuencia1 = pd.DataFrame()
tabla_frecuencia1['ni'] = num_hijos.value_counts()
tabla_frecuencia1['Número de hijos'] = tabla_frecuencia1.index
tabla_frecuencia1.index = range(tabla_frecuencia1['Número de hijos'].count())
tabla_frecuencia1.index.name = 'i'
tabla_frecuencia1['fi(%)'] = (tabla_frecuencia1['ni'] / tabla_frecuencia1['ni'].sum())*100
tabla_frecuencia1['Ni'] = tabla_frecuencia1['ni'].cumsum()
tabla_frecuencia1['Fi(%)'] = tabla_frecuencia1['fi(%)'].cumsum()
tabla_frecuencia1 = tabla_frecuencia1.reindex(columns=['Número de hijos', 'ni', 'fi(%)', 'Ni', 'Fi(%)'])
tabla_frecuencia1
```

**Diagrama de barras**

```python
fig1, ax1 = plt.subplots(nrows=1, ncols=2, figsize = (10,5))
fig1.suptitle('Diagrama de barras', fontsize=32)
ax1[0].bar(x = tabla_frecuencia1['Número de hijos'], height = tabla_frecuencia1['ni'])
ax1[0].set_xlabel('Número de hijos', fontsize=16)
ax1[0].set_ylabel('ni', fontsize=16)
ax1[1].bar(x=tabla_frecuencia1['Número de hijos'], height=tabla_frecuencia1['fi(%)'])
ax1[1].set_xlabel('Número de hijos', fontsize=16)
ax1[1].set_ylabel('fi(%)', fontsize=16)
plt.show()
```

**Diagrama de sectores**

```python
fix, ax2 = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax2[0].pie(tabla_frecuencia1['ni'], labels=tabla_frecuencia1['Número de hijos'], autopct='%.0f%%')
ax2[0].set_xlabel('Número de hijos', fontsize=16)
ax2[1].pie(tabla_frecuencia1['fi(%)'], labels=tabla_frecuencia1['Número de hijos'], autopct='%.0f%%')
ax2[1].set_xlabel('Número de hijos', fontsize=16)
plt.show()
```

**Ejemplo 2 - Variable cualitativa.** Los grupos sanguíneos de una muestra de 30 personas son:

A, B, B, A, AB, 0, 0, A, B, B, A, A, A, A, AB, A, A, A, B, 0, B, B, B, A, A, A, 0, A, AB, 0.

La tabla de frecuencias del grupo sanguíneo en esta muestra es

```python
datos = ['A', 'B', 'B', 'A', 'AB', '0', '0', 'A', 'B', 'B', 'A', 'A', 'A', 'A', 'AB', 'A', 'A', 'A', 'B', '0', 'B', 'B', 'B', 'A', 'A', 'A', '0', 'A', 'AB', '0']
grupo_sanguineo = pd.Series(data=datos)
grupo_sanguineo.head()
```

```python
#Determinando la frecuencia absoluta (ni)
tabla_frecuencia2 = pd.DataFrame()
tabla_frecuencia2['ni'] = grupo_sanguineo.value_counts()
tabla_frecuencia2['Grupo sanguíneo'] = tabla_frecuencia2.index
tabla_frecuencia2.index = range(tabla_frecuencia2['ni'].count())
tabla_frecuencia2.index.name = 'i'
tabla_frecuencia2['fi(%)'] = (tabla_frecuencia2['ni'] / tabla_frecuencia2['ni'].sum()) * 100
tabla_frecuencia2['Ni'] = tabla_frecuencia2['ni'].cumsum()
tabla_frecuencia2['Fi(%)'] = tabla_frecuencia2['fi(%)'].cumsum()
tabla_frecuencia2 = tabla_frecuencia2.reindex(columns=['Grupo sanguíneo', 'ni', 'fi(%)', 'Ni', 'Fi(%)'])
tabla_frecuencia2 = tabla_frecuencia2.round(2)
tabla_frecuencia2
```

**Diagrama de barras ejercicio 2**

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize = (10,5))
fig.suptitle('Diagrama de barras', fontsize=32)
ax[0].bar(x = tabla_frecuencia2['Grupo sanguíneo'], height = tabla_frecuencia2['ni'])
ax[0].set_xlabel('Grupo sanguíneo', fontsize=16)
ax[0].set_ylabel('ni', fontsize=16)
ax[1].bar(x=tabla_frecuencia2['Grupo sanguíneo'], height=tabla_frecuencia2['fi(%)'])
ax[1].set_xlabel('Grupo sanguíneo', fontsize=16)
ax[1].set_ylabel('fi(%)', fontsize=16)
plt.show()
```

**Diagrama de sectores ejercicio 2**

```python
fix, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax[0].pie(tabla_frecuencia2['ni'], labels=tabla_frecuencia2['Grupo sanguíneo'], autopct='%.0f%%')
ax[0].set_xlabel('Grupo sanguíneo', fontsize=16)
ax[1].pie(tabla_frecuencia2['fi(%)'], labels=tabla_frecuencia2['Grupo sanguíneo'], autopct='%.0f%%')
ax[1].set_xlabel('Grupo sanguíneo', fontsize=16)
plt.show()
```
